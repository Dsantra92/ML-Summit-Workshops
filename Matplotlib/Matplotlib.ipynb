# Matplotlib Tutorial 

**Purpose** : To create interesting plots and infographics<br>*Very useful for people learning data analytics*

```python
# Import the matplotlib library
import matplotlib.pyplot as plt

# set the backend of matplotlib to inline
%matplotlib inline

# Import other libraries
import numpy as np
```

## Let's draw our first plot

```python
# x-axis values
x = [1, 3, 5, 6, 7]

# Y-axis values
y = [2, 5, 8, 4, 10]

# Function to plot
plt.plot(x,y)

# function to show the plot
plt.show()
```

These kind of plots are called line plots. They are very useful for understanding sequential data. <br>For example: Plotting the stock market values over a period of time.
<br> *You can play with the x and y values get different plots*

## Exploring different types of plots

Depending upon the data you might want(need) to visualize the data in different ways so that you can extract meaningful infereneces. You might want to visualize different components of the data in many different ways. These matplotlib tools allow you to do just that.

### Scatter Plot

```python
x = range(10)
y = np.random.randn(10)

plt.scatter(x, y)
plt.show()
```

Scatterplots are very useful to understand how the data points is distributed over the space.

### Bar Plot

```python
x = ["car", "bike", "bus", "truck", "bicycle"]
y = [10, 5, 8, 4, 2]

plt.bar(x,y)

plt.xlabel('Item Name') # Set the label for x axis
plt.ylabel('No. of items sold') # Set the label for y axis
plt.title('Autombile Sales') # See the title of the plot

plt.show() 
```

This is more of a real life scenario where we need to visualize the number of sales of different modes of transportation.

*We can try this labelling and titles on all plots. Try this on previous plots.*

### Histograms

```python
# Y-axis values 
y = [11, 11, 11, 15, 18, 2, 3, 3, 3, 1, 3, 4, 2]

plt.hist(y) 

plt.xlabel('quantity')
plt.ylabel('no. of its instances')
plt.title("basic histogram")

plt.show()
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Mathematic graphs")

plt.legend() # Show the labels
plt.show()
```

What if we need to plots of more complicated function ?
> Define your function

```python
# Defining our own function
def arbitary_function(x):
    return  np.power(x, 3) -20 * np.power(x, 2) + 3 * x

x = np.linspace(-10, 20, 100)
y= arbitary_function(x)

plt.plot(x, y)
plt.show()
```

## Customizing plots

### Customizing line plot

```python
x = [1, 3, 5, 6, 7, 10, 12, 15]
y = [2, 5, 8, 5, 10, 9, 11, 12]

# Set the maker point to circle using '-o'. Try changing it to '-s'
# Set the label of the plot -> "Stock Value/hr"
plt.plot(x,y, '-o', label='Stock Value/hr')

plt.legend()
plt.show()
```

### A bit more complicated set of customizations

```python
x = [1, 3, 5, 6, 7, 10, 12, 15]
y = [2, 5, 8, 5, 10, 9, 11, 12]

plt.figure(figsize=(8, 6))

# Set the color of the plot using c
plt.plot(x,y, '-o', label='Stock Value/hr', c = 'r')

plt.xlabel("Hour of the day", fontsize = 16)
plt.ylabel("Stock value", fontsize = 16)
plt.title("Stock exchnage report", fontsize = 16, fontweight = 'bold')

plt.legend(loc='lower right', shadow = True)
plt.show()
```

Matplotlib is not restricted to only these set of customizations. Go to their [official page](https://matplotlib.org/3.3.3/tutorials/index.html) to get a more indepth knowledge of all the funcionalities.  

## Plotting multiple plots together

We can combine two or more plots in primarily 2 different ways.
1. Render them in same plot space
2. Use sperate spaces for different plots

### Combine the plots

```python
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--') # '--' indicates hashed line , you can also try '-.'
plt.show()
```

### Like the plot? Save it on your local machine.

```python
fig.savefig('image.jpeg') # saves the last plot as image.jpeg
```

### Subplots
We basically divide the plotting area into different parts where we draw different plots

```python
plt.figure()  # create a plot figure

x1 = np.linspace(0, 10, 100)

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x1, np.sin(x1), 'blue', label = "sin(x)")
plt.legend(loc='lower right')

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x1, np.cos(x1), 'red', label = "cos(x)")
plt.legend(loc='lower right')

plt.show()
```

## Combining it all together

### Plot 1 : Forced Oscillation

```python
def y1(t):
    return ((3*t**2)-5*t) * np.sin(2*np.pi*t)

t1= np.arange(0.0, 10.0, 0.1)
t2= np.arange(0.0, 10.0, 0.05)

plt.figure(figsize=(20,12))
plt.subplot(211)
plt.plot(t1, y1(t1), 'bo', t2, y1(t2), 'k')

plt.show()
```

### Plot 2 : Scatter plot

```python
x = np.random.randn(1, 100)
y = np.random.randn(1,100)

x1 = np.random.randn(1, 100)
y1 = np.random.randn(1,100)


plt.scatter(x, y, color='blue', s=30, label='0') ## here s is size of point in scatter plot
plt.scatter(x1, y1, color='red', s=30, label='1') 
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend(loc='lower right')
plt.title('Two Examples')
plt.show()
```

### Plot 3: A different way for subplots

```python
# ax will be an array of two Axes objects
_, ax = plt.subplots(4)


x=np.arange(1,10,0.01)
# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
ax[2].plot(x, 3*x**2)
ax[3].plot(x, (7*x**4+5*x**2-50))

plt.show()
```
